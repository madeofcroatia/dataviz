<h2>Pandas</h2>
The primary object of Pandas is the <code>DataFrame</code>. We will use Pandas to read data and store it into <code>DataFrame</code> instance.

```python
df = pd.read_csv('../data/gfmt_sleep.csv', na_values='*')
```

```python
# Look at the contents of first 5 rows:
df.head()
```

<code>DataFrame</code> is always indexed by columns, with column names(<code>string</code> values) being the keys.

```python
df['percent correct'].head()
```

To refer to a particular element of the column we can use integer indexes.

```python
df['percent correct'][4]
```

However, this is <b>not</b> the preferred way to do this. It is better to use <code>loc</code> method:

```python
df.loc[4, 'percent correct']
```

<b>Row indices need not be integers</b>, and we should not count on that. In practice, almost always we will use <b>boolean indexing</b>.

```python
# Get the value(s) in the column 'percent correct' with the value in 'participant number' 42
df.loc[df['participant number'] == 42, 'percent correct']
```

```python
# Get the entire row about participant with number 42
df.loc[df['participant number'] == 42, :]
```

We can use boolean bitwise operators in the boolean indexing. <code>&</code> for bitwise AND, <code>|</code> for bitwise OR, <code>~</code> for bitwise NOT.

```python
# Get entries for women under 21
df.loc[(df['age'] < 21) & (df['gender'] == 'f'), :]
```

```python
# We create three boolean arrays, and for each the entries are True if and only if the
# Boolean statement evaluated at the identical cell at the expression on the right is True
# and False otherwise

under30 = df['age'] < 30
women = df['gender'] == 'f'
good_performers = df['percent correct'] > 85
```

```python
# Get the entries of Women under 30 who performed good on the face matching test
df.loc[under30 & women & good_performers, :]
```

This would normally require us to use loops, but the numpy and pandas libraries allow us to treat all this as simple non-loop operations.

```python
# initialize the list of Boolean indices
indices = [False] * len(df)

# Iterate over the rows of the DataFrame to check if the row should be included
for i, r in df.iterrows():
    if r['age'] < 30 and r['gender'] == 'f' and r['percent correct'] > 85:
        indices[i] = True

#Make our selection with Boolean indexing
df.loc[indices, :]
```

<code>df.iterrows()</code> creates an iterator over the rows of the <code>DataFrame</code>. As we can see, <code>DataFrame</code> is a powerful tool that allows us to do all that looping without worrying about the implementation, while keeping the readability to maximum. The iterator itself is rarely used, but there may be cases when it is needed. As we will see now, most of the times we can just do calculations on <code>DataFrame</code> directly.

<h2>Calculating with data frames</h2>
Test subject is said to have insomnia if they have an SCI of 16 at most. We might want to add a column that explicitly says if the subject has insomnia.

```python
(df['sci'] <= 16) # --- Create a boolean column that is true if and only if the subject for
                  #     whom the row corresponds has insomnia.
```

```python
# Create a column with the name "insomnia" which is the Boolean column cell of which are 
# True if and only if the SCI of the subject is at most 16, and False otherwise.
df['insomnia'] = df['sci'] <= 16

# Taking a peek!
df.head()
```

Applying <code><=</code> operator to a <code>Series</code> resulted in <b>elementwise</b> application. This is called <strong>vectorization</strong>. Just like with numpy arrays, we can apply numpy functions to Pandas objects:

```python
print('Insomniacs:', np.mean(df.loc[df['insomnia'], 'percent correct']))
print('Control:', np.mean(df.loc[~df['insomnia'], 'percent correct']))
```

```python
# DataFrame.describe method computes for us all the summary statistics we usually need
# for each column
df.describe()
```

<code>describe</code> is also a Pandas <code>DataFrame</code>, so we can use indexing, everything we did with <code>DataFrame</code>'s so far on it:

```python
# Get the median for every column in df
df.describe().loc['50%', :]
```

<h2>Slicing with iloc</h2>

Letâs say we wanted to consider only the males in the study. We can slice them out and put them in a new data frame (actually a view into the existing data frame df) using Boolean indexing.

```python
df_males = df.loc[df['gender'] == 'm', :]

df_males.head()
```

Note that the numbers of rows are here from the original <code>df</code>, and not 1, 2, 3, 4, 5... This is actually not too bad, since the row indices need not be integers! There is however a way to index through rows and columns by integers:

```python
# Accessing "third" row and "zeroth" column with iloc
df_males.iloc[3, 0]
```

However, we may be relying here on the <code>DataFrame</code> being sorted in a particular order, which is never a good practice. So, in reality we should never need to use <code>iloc</code>. <strong>Only</strong> use it when the order is irrelevant.

<h2>Renaming columns</h2>

Truth be told, sci, psqi, ess do not tell much us without knowing what these abbreviations stand for. <i>Explicit <b>is</b> better than Implicit</i>, so let us make these column names longer:

```python
# We need to make a renaming dictionary:
rename_dict = {
    "ess": "Epworth Sleepiness Scale",
    "sci": "Sleep Condition Indicator",
    "psqi": "Pittsburgh Sleep Quality Index",
}

# Rename the columns by calling the renaming dictionary
df = df.rename(columns=rename_dict)

df.head()
```

Boolean indexing comes with a payoff, as it is essentially building a boolean array and reverse dictionary look-up for every <code>True</code> value, while non-boolean indexing is just a constant operation.

```python
print("Boolean indexing(slow):")
%timeit df.loc[df['participant number'] == 42, 'percent correct']

print("\nDirect indexing(fast):")
%timeit df.loc[54, 'percent correct']
```
